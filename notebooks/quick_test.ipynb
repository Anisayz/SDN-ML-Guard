```python
import pandas as pd
df = pd.read_csv(r"../data/raw/Friday-02-03-2018_TrafficForML_CICFlowMeter.csv")
```

```python
df = df[df['Label'] != 'Label']
```

```python
(df['Label'] == 'Label').sum()
```

```python
label = df['Label']
df = df.drop(columns=['Label'])

df = df.apply(pd.to_numeric, errors='coerce')

df['Label'] = label
```
